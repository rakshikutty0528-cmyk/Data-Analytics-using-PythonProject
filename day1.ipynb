```python
mark1=int(input("enter mark 1:"))
mark2=int(input("enter mark 2:"))
mark3=int(input("enter mark 3:"))
total=mark1+mark2+mark3
average=total/3
print("total:",total)
print("average:",average)
if average>=50:
    print("pass")
else:
        print("fail")
```

```python
tamil = 90
english = 85
maths = 95
science = 88
social = 92

total = tamil + english + maths + science + social

print("Tamil =", tamil)
print("English =", english)
print("Maths =", maths)
print("Science =", science)
print("Social =", social)

print("Total Marks =", total)

if total >= 450:
    print("Grade : A+")
elif total >= 400:
    print("Grade : A")
elif total >= 350:
    print("Grade : B")
elif total >= 250:
    print("Grade : C")
elif total >= 175:
    print("Grade : Pass")
else:
    print("Grade : Fail")
```
