```python
import matplotlib.pyplot as plt

products = ["Laptop", "Mobile", "Tablet"]
sales = [50, 80, 30]

plt.bar(products, sales, color=["red", "blue", "green"])

plt.title("Product Sales")

plt.show()
```

```python

import matplotlib.pyplot as plt

expenses = [5000, 3000, 2000]

categories = ["Food", "Travel", "Shopping"]

plt.pie(expenses, labels=categories)

plt.show()
```

```python
import matplotlib.pyplot as plt

expenses = [5000, 3000, 2000]

categories = ["Food", "Travel", "Shopping"]

plt.pie(expenses, labels=categories,autopct="%1.1f%%")

plt.show()
```

```python
import matplotlib.pyplot as plt

expenses = [5000, 3000, 2000]

categories = ["Food", "Travel", "Shopping"]

plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%", colors=["pink", "yellow", "green"])

plt.title("Expense Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [10000, 12000, 15000, 13000, 18000, 20000]

plt.figure(figsize=(12, 5))

# Line Chart
plt.subplot(1, 2, 1)

plt.plot(
    months,
    sales,
    marker="o"
)

plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales")

# Bar Chart
plt.subplot(1, 2, 2)

plt.bar(
    months,
    sales,
    color="orange"
)
```

```python
import matplotlib.pyplot as plt

# Dataset: List of student marks
marks = [45, 50, 55, 60, 65, 70, 72, 75, 80, 85, 90, 95]

# Create the histogram
plt.hist(marks)

# Add labels and title
plt.title("Student Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35, 40, 50, 60, 65, 75, 85, 95]

plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
values = [112.5, 115.0, 118.8, 112.5, 116.2, 125.0]
colors = ['#0f3da3', '#d9381e', '#235c33', '#ab1c1c', '#6055cd', '#613644']

# INCREASED FIGURE SIZE AND TEXT SIZES FOR A BIGGER DISPLAY
plt.figure(figsize=(14, 14))

plt.pie(
    values,
    labels=months,
    colors=colors,
    autopct='%1.1f%%',
    textprops={'fontsize': 18}  # Makes the text font much larger
)

plt.show()
```
