Practice : Employee dataset exploration

```python
import pandas as pd

# Create employee dataset
data = {
    "Name": ["Anitha", "Priya", "Karthik", "Divya"],
    "Salary": [30000, 45000, 40000, 50000]
}

df = pd.DataFrame(data)

# Display dataset
print("Employee Dataset")
print(df)

# Total employees
print("\nTotal Employees:", len(df))

# Average salary
print("Average Salary:", df["Salary"].mean())

# Highest salary
print("Highest Salary:", df["Salary"].max())
```
