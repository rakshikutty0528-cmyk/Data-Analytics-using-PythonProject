```python
# =====================================================
# HR Employee Attrition Analysis Using Python
# =====================================================

# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Create HR Dataset
# ----------------------------
data = {
    "Employee_ID":[101,102,103,104,105,106,107,108,109,110],
    "Department":["HR","IT","Finance","IT","HR","Sales","Sales","IT","Finance","HR"],
    "Salary":[25000,45000,35000,50000,28000,30000,32000,48000,36000,27000],
    "Experience":[2,5,4,6,1,3,4,7,5,2],
    "Attrition":["Yes","No","Yes","No","Yes","No","Yes","No","Yes","Yes"]
}

# ----------------------------
# Step 2: Create DataFrame
# ----------------------------
df = pd.DataFrame(data)

print("="*60)
print("        HR EMPLOYEE DATASET")
print("="*60)
print(df)

# ----------------------------
# Step 3: Data Cleaning
# ----------------------------
print("\nChecking Missing Values")
print(df.isnull().sum())

print("\nRemoving Duplicate Records...")
df = df.drop_duplicates()

# ----------------------------
# Step 4: Basic Information
# ----------------------------
print("\nDataset Information")
print(df.info())

print("\nDataset Description")
print(df.describe())

# ----------------------------
# Step 5: Employee Attrition Analysis
# ----------------------------
total_employees = len(df)
employees_left = len(df[df["Attrition"]=="Yes"])
employees_stayed = len(df[df["Attrition"]=="No"])

attrition_rate = (employees_left/total_employees)*100

print("\n==============================")
print("EMPLOYEE ATTRITION ANALYSIS")
print("==============================")
print("Total Employees :", total_employees)
print("Employees Left :", employees_left)
print("Employees Stayed :", employees_stayed)
print("Employee Attrition Rate : {:.2f}%".format(attrition_rate))

# ----------------------------
# Step 6: Department-wise Attrition
# ----------------------------
dept_attrition = df.groupby("Department")["Attrition"].apply(
    lambda x:(x=="Yes").sum()
)

print("\nDepartment-wise Attrition")
print(dept_attrition)

# ----------------------------
# Step 7: Average Salary
# ----------------------------
salary_analysis = df.groupby("Department")["Salary"].mean()

print("\nAverage Salary by Department")
print(salary_analysis)

# ----------------------------
# Step 8: Average Experience
# ----------------------------
experience_analysis = df.groupby("Department")["Experience"].mean()

print("\nAverage Experience by Department")
print(experience_analysis)

# ----------------------------
# Step 9: Employees Who Left
# ----------------------------
print("\nEmployees Who Left")
print(df[df["Attrition"]=="Yes"])

# =====================================================
# DATA VISUALIZATION
# =====================================================

# Pie Chart
plt.figure(figsize=(6,6))
plt.pie(
    [employees_left, employees_stayed],
    labels=["Left","Stayed"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Employee Attrition Distribution")
plt.show()

# Department-wise Attrition
plt.figure(figsize=(7,5))
dept_attrition.plot(kind="bar")
plt.title("Department-wise Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Employees Left")
plt.grid(True)
plt.show()

# Average Salary
plt.figure(figsize=(7,5))
salary_analysis.plot(kind="bar")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.grid(True)
plt.show()

# Average Experience
plt.figure(figsize=(7,5))
experience_analysis.plot(kind="bar")
plt.title("Average Experience by Department")
plt.xlabel("Department")
plt.ylabel("Average Experience")
plt.grid(True)
plt.show()

print("\n========================================")
print("HR Employee Attrition Analysis Completed")
print("========================================")
```
