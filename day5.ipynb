Practice 1 : Product sales performance analysis

```python
import pandas as pd

# Create product sales dataset
data = {
    "Product": ["Pen", "Pencil", "Book", "Eraser"],
    "Sales": [200, 150, 500, 100]
}

df = pd.DataFrame(data)

# Display dataset
print("Product Sales Dataset")
print(df)

# Total sales
print("\nTotal Sales:", df["Sales"].sum())

# Average sales
print("Average Sales:", df["Sales"].mean())

# Highest sales
print("Highest Sales:", df["Sales"].max())
```
